```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder 

labelEncoder = LabelEncoder()
oneHotEncoder = OneHotEncoder(sparse_output = False)

one_hot_columns = ['Leather interior', 'Gear box type', 'Drive wheels', 'Wheel']
label_encode_columns = ['Manufacturer', 'Model', 'Category', 'Fuel type', 'Color']
```

```python
label_encoders = {}
for column in label_encode_columns:
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
    label_encoders[column] = label_encoder
```

```python
label_encoders
```

```python
# Save the label encoders for future use
with open('../models/label_encoders.pkl', 'wb') as f:
    pickle.dump(label_encoders, f)
```

```python
data.head()
```

```python
one_hot_encoder_array = oneHotEncoder.fit_transform(data[one_hot_columns])

one_hot_encoder_df = pd.DataFrame(one_hot_encoder_array , columns=oneHotEncoder.get_feature_names_out(one_hot_columns) , index = data.index)

data = pd.concat([data , one_hot_encoder_df] , axis=1)
data.drop(one_hot_columns ,axis=1, inplace=True)
data
```

```python
# Save the encoder for future use
with open('../models/one_hot_encoder.pkl', 'wb') as f:
    pickle.dump(oneHotEncoder, f)
```

```python
X = data.drop(['Price'] , axis=1)
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.30 , random_state=42)

print(f"Train set: {len(X_train)} samples")
print(f"Test set: {len(X_test)} samples")                                                   
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Save the scaler
with open('../models/scaler.pkl', 'wb') as f:
    pickle.dump(sc, f)
```

```python
from sklearn.ensemble import RandomForestRegressor 

randomForestModel = RandomForestRegressor()

randomForestModel.fit(X_train , y_train)
```

```python
y_pred = randomForestModel.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error ,r2_score
rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Root Mean Squared Error: {rmse}")
print(f"R^2 Score: {r2}")
```
